```python
#loading image from url
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from skimage import io
from skimage.transform import rescale
from skimage import  color
import scipy.ndimage as ndi
url = "https://upload.wikimedia.org/wikipedia/commons/thumb/a/a4/Lat_lebka.jpg/330px-Lat_lebka.jpg"
#url = "https://upload.wikimedia.org/wikipedia/commons/thumb/d/d0/AP_lumbar_xray.jpg/255px-AP_lumbar_xray.jpg"
a = io.imread(url)

im = color.rgb2gray(a)
#rescaling
a = image_rescaled = rescale(im, 0.5, anti_aliasing=False)
plt.imshow(a, cmap = "gray")
plt.axis('off')
plt.show()
```

```python
from skimage.filters import threshold_otsu
#otsu's thresholding

image = im
thresh = threshold_otsu(image)
binary = image > thresh

fig, axes = plt.subplots(ncols=3, figsize=(8, 2.5))
ax = axes.ravel()
ax[0] = plt.subplot(1, 2, 1, adjustable='box')
ax[1] = plt.subplot(1, 2, 2, sharex=ax[0], sharey=ax[0], adjustable='box')

ax[0].imshow(image, cmap=plt.cm.gray)
ax[0].set_title('Original')
ax[0].axis('off')

ax[1].imshow(binary, cmap=plt.cm.gray)
ax[1].set_title('Thresholded')
ax[1].axis('off')

plt.show()
```

```python
#dilatation
from scipy import ndimage
from skimage.filters import threshold_otsu
from skimage.morphology import disk
from skimage.morphology import erosion

fig, axes = plt.subplots(ncols=4, nrows = 1, sharex=True, sharey=True,
                         figsize=(12, 5))

image = im
thresh = threshold_otsu(image)
binary = image > thresh

dilated = ndimage.binary_dilation(binary, structure=np.ones((3,3)))
dilated1 = ndimage.binary_dilation(binary, structure=np.ones((5,5)))
dilated2 = ndimage.binary_dilation(binary, structure=np.ones((9,9)))

axes[0].imshow(image, cmap=plt.cm.gray)
axes[0].set_title('original image')
axes[1].imshow(dilated, cmap=plt.cm.gray)
axes[1].set_title('dilatation kernel: 3x3')
axes[2].imshow(dilated1, cmap=plt.cm.gray)
axes[2].set_title('dilatation kernel: 5x5')
axes[3].imshow(dilated2, cmap=plt.cm.gray)
axes[3].set_title('dilatation kernel: 9x9')
```

```python
#erosion

fig, axes = plt.subplots(ncols=4, nrows = 1, sharex=True, sharey=True,
                         figsize=(12, 5))

image = im
thresh = threshold_otsu(image)
binary = image > thresh

eroded = ndimage.binary_erosion(binary, structure=np.ones((3,3)))
eroded1 = ndimage.binary_erosion(binary, structure=np.ones((5,5)))
eroded2 = ndimage.binary_erosion(binary, structure=np.ones((9,9)))

axes[0].imshow(image, cmap=plt.cm.gray)
axes[0].set_title('original image')
axes[1].imshow(eroded, cmap=plt.cm.gray)
axes[1].set_title('erosion kernel: 3x3')
axes[2].imshow(eroded1, cmap=plt.cm.gray)
axes[2].set_title('erosion kernel: 5x5')
axes[3].imshow(eroded2, cmap=plt.cm.gray)
axes[3].set_title('erosion kernel: 9x9')
```

```python
#closing

fig, axes = plt.subplots(ncols=4, nrows = 1, sharex=True, sharey=True,
                         figsize=(12, 5))

image = im
thresh = threshold_otsu(image)
binary = image > thresh

closed = ndimage.binary_closing(binary, structure=np.ones((3,3)))
closed1 = ndimage.binary_closing(binary, structure=np.ones((5,5)))
closed2 = ndimage.binary_closing(binary, structure=np.ones((9,9)))

axes[0].imshow(image, cmap=plt.cm.gray)
axes[0].set_title('original image')
axes[1].imshow(closed, cmap=plt.cm.gray)
axes[1].set_title('closing kernel: 3x3')
axes[2].imshow(closed1, cmap=plt.cm.gray)
axes[2].set_title('closing kernel: 5x5')
axes[3].imshow(closed2, cmap=plt.cm.gray)
axes[3].set_title('closing kernel: 9x9')
```

```python
#opening

fig, axes = plt.subplots(ncols=4, nrows = 1, sharex=True, sharey=True,
                         figsize=(12, 5))

image = im
thresh = threshold_otsu(image)
binary = image > thresh

opened = ndimage.binary_opening(binary, structure=np.ones((3,3)))
opened1 = ndimage.binary_opening(binary, structure=np.ones((5,5)))
opened2 = ndimage.binary_opening(binary, structure=np.ones((9,9)))

axes[0].imshow(image, cmap=plt.cm.gray)
axes[0].set_title('original image')
axes[1].imshow(opened, cmap=plt.cm.gray)
axes[1].set_title('opening kernel: 3x3')
axes[2].imshow(opened1, cmap=plt.cm.gray)
axes[2].set_title('opening kernel: 5x5')
axes[3].imshow(opened2, cmap=plt.cm.gray)
axes[3].set_title('opening kernel: 9x9')
```

```python
from scipy import ndimage
from skimage.filters import threshold_otsu
from skimage.morphology import disk
from skimage.morphology import erosion

image = im
thresh = threshold_otsu(image)
binary = image > thresh
eroded = ndimage.binary_erosion(binary, structure=np.ones((5,5)))
dilated = ndimage.binary_dilation(binary, structure=np.ones((5,5)))
opening = ndimage.binary_opening(binary, structure=np.ones((5,5)))
closing = ndimage.binary_closing(binary, structure=np.ones((5,5)))

fig, axes = plt.subplots(ncols=3, nrows = 2, sharex=True, sharey=True,
                         figsize=(12, 8))

axes[0][0].imshow(image, cmap=plt.cm.gray)
axes[0][0].set_title('original image')

axes[0][1].imshow(binary, cmap=plt.cm.gray)
axes[0][1].set_title('otsu threshold')

axes[0][2].imshow(eroded, cmap=plt.cm.gray)
axes[0][2].set_title('Binary erosion')

axes[1][0].imshow(dilated, cmap=plt.cm.gray)
axes[1][0].set_title('Binary dilatation')

axes[1][1].imshow(opening, cmap=plt.cm.gray)
axes[1][1].set_title('Binary opening')

axes[1][2].imshow(closing, cmap=plt.cm.gray)
axes[1][2].set_title('Binary closing')

for ax in axes.ravel():
    ax.axis('off')

plt.tight_layout()
```

```python
fig, axes = plt.subplots(ncols=4, nrows = 1, sharex=True, sharey=True,
                         figsize=(12, 5))

image = im
thresh = threshold_otsu(image)
binary = image > thresh

eroded = ndimage.binary_erosion(binary, structure=np.ones((3,3)))
boundary =binary ^ eroded

axes[0].imshow(image, cmap=plt.cm.gray)
axes[0].set_title('original image')
axes[1].imshow(binary, cmap=plt.cm.gray)
axes[1].set_title('otsu threshold')
axes[2].imshow(eroded, cmap=plt.cm.gray)
axes[2].set_title('eroded')
axes[3].imshow(boundary, cmap=plt.cm.gray)
axes[3].set_title('boundary detection')
```

```python

url = "https://upload.wikimedia.org/wikipedia/commons/thumb/6/6d/H4IIE_cells_with_stained_nuclei.jpg/640px-H4IIE_cells_with_stained_nuclei.jpg"

a = io.imread(url)

#we want to select only the nucleus
#thus only what is blue
im = a[:,:,2]
image = im
thresh = threshold_otsu(image)
binary = image > thresh
eroded = ndimage.binary_erosion(binary, structure=np.ones((7,7)))
opening = ndimage.binary_opening(eroded, structure=np.ones((11,11)))
boundary =binary ^ opening

fig, axes = plt.subplots(ncols=3, figsize=(8, 2.5))
ax = axes.ravel()
ax[0] = plt.subplot(1, 2, 1, adjustable='box')
ax[1] = plt.subplot(1, 2, 2, sharex=ax[0], sharey=ax[0], adjustable='box')

ax[0].imshow(a)
ax[0].set_title('Original')
ax[0].axis('off')

ax[1].imshow(boundary, cmap=plt.cm.gray)
ax[1].set_title('morphology operations')
ax[1].axis('off')

plt.show()
```

```python
url = "https://images.unsplash.com/photo-1599388050944-ab3cc1c76883?ixlib=rb-1.2.1&ixid=MnwxMjA3fDB8MHxwaG90by1wYWdlfHx8fGVufDB8fHx8&auto=format&fit=crop&w=423&q=80"
a = io.imread(url)

im = color.rgb2gray(a)

image = im
thresh = threshold_otsu(image)
binary = image > thresh
opening = ndimage.binary_opening(binary, structure=np.ones((13,5)))
closing = ndimage.binary_closing(binary, structure=np.ones((13,5)))
black_hat =  im - closing
white_hat = im - opening

fig, axes = plt.subplots(ncols=3, figsize=(12, 6))
ax = axes.ravel()
ax[0] = plt.subplot(1, 2, 1, adjustable='box')
ax[1] = plt.subplot(1, 2, 2, sharex=ax[0], sharey=ax[0], adjustable='box')

ax[0].imshow(a)
ax[0].set_title('Original')
ax[0].axis('off')

ax[1].imshow(black_hat, cmap=plt.cm.gray)
ax[1].set_title('morphology operations')
ax[1].axis('off')

plt.show()
```
