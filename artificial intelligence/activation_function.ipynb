```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x =np.linspace(-10, 10, num = 100) 
```

```python
def linear(x, a =2):
    return a*x

def linear_dash(x, a =2):
    f = np.linspace(1, 1, len(x)) 
    return f* a

#plotting the function and the derivative
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
fig.subplots_adjust(hspace = .1, wspace=0.5)
axs = axes.ravel()

y =linear(x)
y_dash =linear_dash(x)
axs[0].plot(x, y, label= 'linear function', linewidth=3)
axs[0].axvline(x=0, color='k', linestyle='--')
axs[0].axhline(y=0, color='k', linestyle='--')
axs[0].set_xlabel('Input value')
axs[0].set_ylabel('Output value')
axs[0].set_title('activation function')
axs[0].legend(loc='lower right')
axs[1].plot(x, y_dash, label='derivative', color = 'red', linewidth=3)
axs[1].axvline(x=0, color='k', linestyle='--')
axs[1].axhline(y=0, color='k', linestyle='--')
axs[1].set_xlabel('Input value')
axs[1].set_ylabel('Output value')
axs[1].set_title('derivative')
axs[1].legend(loc='lower right')
axs[1].set_ylim([-2.5, 2.5])
```

```python
def binary_step(x, a =2):
  return np.where(x<=0,0,1)

def binary_step_dash( a=0):
    return np.where(x<=0,0,0)

#plotting the function and the derivative
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
fig.subplots_adjust(hspace = .1, wspace=0.5)
axs = axes.ravel()

y =binary_step(x)
y_dash =binary_step_dash(x)
axs[0].plot(x, y, label= 'Binary step', linewidth=3)
axs[0].axvline(x=0, color='k', linestyle='--')
axs[0].axhline(y=0, color='k', linestyle='--')
axs[0].set_xlabel('Input value')
axs[0].set_ylabel('Output value')
axs[0].set_title('activation function')
axs[0].legend(loc='lower right')
axs[1].plot(x, y_dash, label='derivative', color = 'red', linewidth=3)
axs[1].axvline(x=0, color='k', linestyle='--')
axs[1].axhline(y=0, color='k', linestyle='--')
axs[1].set_xlabel('Input value')
axs[1].set_ylabel('Output value')
axs[1].set_title('derivative')
axs[1].legend(loc='lower right')
axs[1].set_ylim([-2.5, 2.5])
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))

def sigmoid_dash(x):
    return sigmoid(x)*(1-sigmoid(x))

#plotting the function and the derivative
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
fig.subplots_adjust(hspace = .1, wspace=0.5)
axs = axes.ravel()

y =sigmoid(x)
y_dash =sigmoid_dash(x)
axs[0].plot(x, y, label= 'Sigmoid/logistic', linewidth=3)
axs[0].axvline(x=0, color='k', linestyle='--')
axs[0].axhline(y=0, color='k', linestyle='--')
axs[0].set_xlabel('Input value')
axs[0].set_ylabel('Output value')
axs[0].set_title('activation function')
axs[0].legend(loc='lower right')
axs[1].plot(x, y_dash, label='derivative', color = 'red', linewidth=3)
axs[1].axvline(x=0, color='k', linestyle='--')
axs[1].axhline(y=0, color='k', linestyle='--')
axs[1].set_xlabel('Input value')
axs[1].set_ylabel('Output value')
axs[1].set_title('derivative')
axs[1].legend(loc='lower right')
axs[1].set_ylim([-2.5, 2.5])
```

```python
def tanh(x):
  return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))

def tanh_dash(x):
    return 1-(tanh(x))**2

#plotting the function and the derivative
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
fig.subplots_adjust(hspace = .1, wspace=0.5)
axs = axes.ravel()

y =tanh(x)
y_dash =tanh_dash(x)
axs[0].plot(x, y, label= 'Tanh', linewidth=3)
axs[0].axvline(x=0, color='k', linestyle='--')
axs[0].axhline(y=0, color='k', linestyle='--')
axs[0].set_xlabel('Input value')
axs[0].set_ylabel('Output value')
axs[0].set_title('activation function')
axs[0].legend(loc='lower right')
axs[1].plot(x, y_dash, label='derivative', color = 'red', linewidth=3)
axs[1].axvline(x=0, color='k', linestyle='--')
axs[1].axhline(y=0, color='k', linestyle='--')
axs[1].set_xlabel('Input value')
axs[1].set_ylabel('Output value')
axs[1].set_title('derivative')
axs[1].legend(loc='lower right')
axs[1].set_ylim([-2.5, 2.5])
```
